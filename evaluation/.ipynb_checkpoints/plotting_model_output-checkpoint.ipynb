## Below is a cell for the imports and legends

- "genre" is a variable if we're doing this for a specific genre
- the "xxx_patch" are the variables that are used in the plot for the senses, they match the colours used below
- "colours" is a list of colours. They should be in the same order as the senses (xxx_patch)
- "target" is the target word.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import matplotlib
matplotlib.rcParams['figure.figsize'] = [10,7]
import matplotlib.patches as mpatches
import os
import io
import re



dir_in = os.path.abspath(os.path.join(os.path.dirname(os.getcwd()), "evaluation", "plotting_test"))
print("We're working with output files in",dir_in)

k_senses = io.open(dir_in+"/example_3genre_output.dat_1","r")


## Put the genre here, for filesaving purposes among other things
genre = os.path.basename(k_senses.name)
genre = genre.split(".")[1]
genre = genre.replace("dat","")
genre = "genre"+genre 



## change this for target word. It can't be read from file
target = "kosmos"

print("We're working on target word",target,"and with the genre",genre)

```

## Below is a cell where we read the output file

```python
output_senses = k_senses.read()

lines_output = output_senses.split("===============  per time  ===============")[0].split("\n")
number_of_the_k = 0
k_and_words = dict()

#### Here we're reading the output file to get the top words, for the plot labels.
for line in lines_output:
    if line[:6] == "p(w|s)":
        
        word_list = list()
        line = line.split(":")[1]
        line = line.split(";")
        #print(number_of_the_k,line)
        
        line = line[:-1] # last item of the list is empty
        for word_prob in line:
            word_prob = word_prob.lstrip()
            word = word_prob.split(" ")[0]
            word_list.append(word)
            
        k_and_words[number_of_the_k] = word_list   
        number_of_the_k += 1
print(k_and_words)



#### Here we're reading the output file to get the probabilities for each period.

lines_output_plot = output_senses.split("===============  per time  ===============")[1].split("\n")
period_relative_model = dict()

for i in range(0,len(lines_output_plot)):
    if lines_output_plot[i][0:5] == "Time=":  # if a line starts with "time" we take it into account
        for x in range(i,i+number_of_the_k+1): # for every "number of  the k" lines that follow
        
            if lines_output_plot[x][0:5] == "Time=": # if a line starts with "time" we take the value for the slice
                #print(lines_output_plot[x][0:15])
                string = lines_output_plot[x][0:15].replace("Time=","")
                string = string.replace("  ","")
                period = string
                templist = list()
                
            if lines_output_plot[x][0:5] != "Time=":  # if a line doesn't start with "time" but is considered(cf line3)
                ligne = re.split("\s{3,}",lines_output_plot[x]) # we take the first part of the line (importance of that K)
                

                templist.append(float(ligne[0]))
                
        period_relative_model[str(period)] = templist
                
                
print(period_relative_model)
```
