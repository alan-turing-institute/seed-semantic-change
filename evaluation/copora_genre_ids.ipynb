```python
# Basic variables and imports:

import codecs, csv, os, time, re, io
from os import listdir
from os.path import isfile, join
from  more_itertools import unique_everseen

# directories

dir_in = os.path.abspath(os.path.join(os.path.dirname(os.getcwd()), "src", "dynamic-senses","greek_input","all_results"))
dir_out = os.path.abspath(os.path.join(os.path.dirname(os.getcwd()), "src", "dynamic-senses","greek_input","simon_genre_corpora"))

dir_expert = os.path.abspath(os.path.join(os.path.dirname(os.getcwd()), "corpus_scripts_output")) 
```

```python
print(dir_in)
print(dir_out)
print(dir_expert)
```

```python
liste_genres_files = os.listdir(dir_expert)
liste_genres_files_true = list()
for fichier in liste_genres_files:
    if fichier[0:2] == "g_":
        if fichier[-7:] == "ids.txt":
            liste_genres_files_true.append(fichier)
```

```python
print(liste_genres_files_true)
```

```python
dic_of_genres = dict()
dic_of_genres["Comedy"] = 0
dic_of_genres["Essays"] = 1
dic_of_genres["Letters"] = 2
dic_of_genres["Narrative"] = 3
dic_of_genres["Oratory"] = 4
dic_of_genres["Philosophy"] = 5
dic_of_genres["Poetry"] = 6
dic_of_genres["Religion"] = 7
dic_of_genres["Technical"] = 8
dic_of_genres["Tragedy"] = 9
```

```python
for fichier in liste_genres_files_true:
    print(fichier[:-4])
```

```python


for fichier in liste_genres_files_true:
    final_mus = io.open(dir_out+"/genre_"+fichier[:-4]+"_mus.txt","w")
    final_harmonia = io.open(dir_out+"/genre_"+fichier[:-4]+"_harmonia.txt","w")
    final_kosmos = io.open(dir_out+"/genre_"+fichier[:-4]+"_kosmos.txt","w")
    original = io.open(dir_expert+"/"+fichier,"r")
    fichier_stripped = fichier.replace("g_","")
    fichier_stripped = fichier_stripped.replace("_ids.txt","")
    for key in dic_of_genres.keys():
        print(key)
        print(fichier_stripped)
        if fichier_stripped == key:
            for line in original.readlines():
                ligne = line.split("\t")
                try:
                    if int(ligne[0]) > -801 and int(ligne[0]) < 143:
                        final_kosmos.write("%s\t%s\t%s" %(ligne[0],dic_of_genres[key],ligne[1]))
                        
                    if int(ligne[0]) > -801 and int(ligne[0]) < 401:
                        final_harmonia.write("%s\t%s\t%s" %(ligne[0],dic_of_genres[key],ligne[1]))
                        
                    if int(ligne[0]) > -426 and int(ligne[0]) < 366:
                        final_mus.write("%s\t%s\t%s" %(ligne[0],dic_of_genres[key],ligne[1]))  

                except IndexError:
                    print("this line fails",line,key)
    
```
